```python
import requests
import pandas as pd
import plotly.express as px

print("🌍 Earthquake Visualizer ready!")
```

```python
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/2.5_day.geojson"
response = requests.get(url)
eq_data = response.json()

print(f"📊 {len(eq_data['features'])} quakes in last 24hrs")
print("Sample:", eq_data['features'][0]['properties']['title'])
```

```python
# Extract data
quakes = []
for feature in eq_data['features']:
    props = feature['properties']
    geom = feature['geometry']['coordinates']  # [lon, lat, depth]
    quakes.append({
        'mag': props['mag'],
        'lat': geom[1],
        'lon': geom[0],
        'place': props['place']
    })

df = pd.DataFrame(quakes)

# Plot
fig = px.scatter_geo(df, lat='lat', lon='lon', size='mag', color='mag',
                     hover_name='place', projection="natural earth",
                     title="Global Earthquakes (Last 24hrs, Mag 2.5+)")
fig.show()

fig.write_image("../figures/global_earthquakes.png", width=1400)
print("✅ PNG saved!")
print(f"💥 Largest: {df['mag'].max():.1f} ({df.loc[df['mag'].idxmax(), 'place']})")
```
